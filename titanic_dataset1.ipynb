```python
sns.distplot(tit['age'])
```

```python
sns.relplot(data=tit, x='who', y='age',hue="who",size="deck")
```

```python
sns.scatterplot(data=tit, x='who', y='age',hue="who",size="deck")
```

```python
sns.stripplot(data=tit,x='who', y='age',hue="alive")
```

```python
sns.swarmplot(data=tit,x='who', y='age',hue="alive")
```

```python
sns.boxplot(tit['pclass'])
```

```python
sns.catplot(data=tit,x='who', y='age',hue="alive",kind="box")
```

```python
sns.catplot(data=tit,x='who', y='age',hue="alive",kind="violin")
```

```python
sns.violinplot(tit['age'])
```

```python
sns.lmplot(x='pclass', y='age', data=tit, hue='alive', markers=True,aspect=2)
```

```python
sns.set_style('darkgrid')
```

```python
sns.jointplot(x='survived', y='pclass', data=tit, hue='who')
```

```python
sns.jointplot(x='who', y='age', data=tit, hue='alive',kind="hist")
```

```python
sns.jointplot(x='survived', y='pclass', data=tit, kind="reg", color="green")
```

```python
t = sns.FacetGrid(tit, col='class')
t.map(sns.histplot, 'who')
```

```python
t = sns.FacetGrid(tit, col='class')
t.map_dataframe(sns.histplot, x='who',color="red")
```

```python
t = sns.FacetGrid(tit, col='class')
t.map_dataframe(sns.scatterplot, x='who',y="age",color="red")
t.set_axis_labels('who', 'age')
```
